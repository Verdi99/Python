## Drop unnecessary columns from the dataset

```python
df.drop(columns=['id','imdb_id','cast','homepage','tagline','overview','budget_adj','revenue_adj'],inplace=True)
```

```python
df.head()
```

## Drop NULL values in the datasets

```python
df.dropna(how='any',subset = ['genres','director','keywords'], inplace=True)
```

```python
df['production_companies'] = df['production_companies'].fillna(0)
```

```python
df.isna().sum()
```

## Creating additional columns (ROI, Profit)

```python
df.insert(3,'profit',df['revenue']-df['budget'])
```

```python
df.insert(4,'roi',df['profit']/df['budget'])
```

```python
df['roi'] = df['roi'].fillna(0)
```

```python
df['roi'].replace(np.inf,0,inplace=True)
```

```python
df['roi'].max()
```

```python
df.head()
```

## Create histogram for specific columns

## The average of return on investment over different release years

```python
df_1 = df.groupby('release_year')['roi'].mean()
plt.figure(figsize=(10,8))
plt.title("The average of return on investment over different years",fontsize=14)
plt.xlabel("Release years of movies",fontsize=12)
plt.ylabel("Return on investment",fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend()
df_1.plot()
```

## Total popularity over years

```python
df_2 = df.groupby("release_year")['popularity'].sum()
plt.figure(figsize=(10,8))
plt.title("Total popularity over different years",fontsize=14)
plt.xlabel("Release years of movies",fontsize=12)
plt.ylabel("Total popularity",fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
df_2.plot()
```

## Average of vote average over different release years 

```python
df_3 = df.groupby("release_year")['vote_average'].mean()
plt.figure(figsize=(10,8))
plt.title("Average of vote average over different release years",fontsize=14)
plt.xlabel("Release years of movies",fontsize=12)
plt.ylabel("Average of vote average",fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
df_3.plot()
```

## Dependency between popularity and vote average (scatter plot)

```python
df_4 = pd.DataFrame(df.groupby("popularity")['vote_average'].mean())
plt.figure(figsize=(10,8))
plt.title("Dependency between popularity and vote average",fontsize=14)
plt.ylabel("Popularity",fontsize=12)
plt.xlabel("Vote average",fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.scatter(x=df_4['vote_average'], y = df_4.index)
```

## Splitting values in genres columns based on a delimiter ("|")

```python
df["genres"] = df["genres"].apply(lambda x: x.split("|"))
```

```python
df
```

```python
df_x = df.explode("genres")
```

## Popularity of movies by genres

```python
df_5 = df_x.groupby("genres")['popularity'].mean().sort_values(ascending=True)
plt.figure(figsize=(10,8))
plt.title("Dependency between genres and popularity",fontsize=14)
plt.xlabel("Popularity",fontsize=12)
plt.ylabel("Genres",fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
df_5.plot(kind="barh")
```

## Setting release_date into date format

```python
df["release_date"] = pd.to_datetime(df["release_date"],errors="coerce")
```

```python
df["release_date"] = df["release_date"].apply(lambda x: pd.Timestamp(x.year-100,x.month,x.day) if x.year>2024 else pd.Timestamp(x.year,x.month,x.day))
```

```python
df["release_date"].dt.year.values
```

## Popular movies by months

```python
df_6 = df.groupby(df["release_date"].dt.month)['popularity'].mean()
plt.figure(figsize=(10,8))
plt.title("Dependency between release month and popularity",fontsize=14)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
df_6.plot.barh(xlabel = "Popularity",ylabel="Month")
```

## Popular movies by revenue

```python
df_7 = df.sort_values(by="revenue",ascending=False)
```

```python
df_71 = df_7.head(20).sort_values(by="popularity")
```

```python
plt.figure(figsize=(10, 8))
df_71.set_index('original_title').plot.barh(y='popularity',legend=False,color="purple")
plt.title("Popularity of Movies Ranked with the Highest Revenue", fontsize=14)
plt.xlabel("Movies", fontsize=12)
plt.ylabel("Popularity", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

## Popular titles by profit

```python
df_8 = df.sort_values(by="profit",ascending=False)
df_81 = df_8.head(20).sort_values(by="popularity",ascending=True)
```

```python
plt.figure(figsize=(10, 8))
df_81.set_index('original_title').plot.barh(y='popularity',legend=False,color="purple")
plt.title("Popularity of Movies Ranked with the Highest Profit", fontsize=14)
plt.xlabel("Movies", fontsize=12)
plt.ylabel("Profit", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

## Top 5 movies by profit (pie chart)

```python
df_9 = df.sort_values(by="profit",ascending=False).head(5)
plt.pie(df_9["profit"],labels = df_9["original_title"],autopct = '%1.1f%%',startangle=90)
plt.title("Most Profitable Movies")
plt.show()
```

## Visualizing popular prodution companies
